```python
# Gradient Boosting in python using scikit-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import ensemble
from sklearn import linear_model

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.arange(0,60)
y = map(lambda x: x / 2 + (x // 10) % 2 * 20 * x / 5 + np.random.random() * 10, x)
x = pd.DataFrame({'x': x})
y = pd.DataFrame(y)

# plot mock data
plt.figure(figsize=(10,5));plt.scatter(x,y);plt.xlabel("x");plt.ylabel("y");plt.show()
```

```python
linear_regressor = linear_model.LinearRegression()

linear_regressor.fit(x,y)

plt.figure(figsize=(10,5));plt.title("Linear Regression");plt.scatter(x,y)
plt.plot(x,linear_regressor.predict(x),color ='r',label="Regression line")
plt.xlabel("x");plt.ylabel("y");plt.legend();plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Residuals $(y - \hat{y})$ of Linear Regression model")
plt.scatter(x,y -linear_regressor.predict(y),color = 'brown')
plt.xlabel("x");plt.ylabel("y");plt.show()
```

```python
# Using Gradient Boosting
# Gradient boosting uses a set of decision trees in series in an ensemble to predict y.
```

```python
gradient_boosting_regressor = ensemble.GradientBoostingRegressor(n_estimators=1,max_depth=1,learning_rate=1)
gradient_boosting_regressor.fit(x,y)

plt.figure(figsize=(10, 5))
plt.title('Gradient Boosting model (1 estimators, Single tree split)')
plt.scatter(x, y);plt.plot(x, gradient_boosting_regressor.predict(x), color='r');plt.xlabel("x");plt.ylabel("y");plt.show()
```

```python
gradient_boosting_regressor = ensemble.GradientBoostingRegressor(n_estimators=2,max_depth=1,learning_rate=1)
gradient_boosting_regressor.fit(x,y)

plt.figure(figsize=(10, 5))
plt.title('Gradient Boosting model (2 estimators, Single tree split)')
plt.scatter(x, y);plt.plot(x, gradient_boosting_regressor.predict(x), color='r');plt.xlabel("x");plt.ylabel("y");plt.show()
```

```python
f,ax = plt.subplots(2,2,figsize =(15,10))

for idx,n_estis in enumerate([5,10,15,20]):
    gradient_boosting_regressor = ensemble.GradientBoostingRegressor(n_estimators=n_estis,max_depth=1,learning_rate=1)
    
    gradient_boosting_regressor.fit(x,y)
    subplot = ax[idx // 2][idx % 2]
    subplot.set_title("Gradient Boosting Model with n_estimator of {}".format(n_estis))
    subplot.scatter(x, y)
    subplot.plot(x, gradient_boosting_regressor.predict(x), color='r')
plt.show()
```

```python
# Plot the graph of n_estimator = 10 and different values of max_depth

f,ax = plt.subplots(2,2,figsize=(15,10))

for idx,max_dept in enumerate([1,2,3,5]):
    gradient_boosting_regressor = ensemble.GradientBoostingRegressor(n_estimators= 10, max_depth = max_dept,learning_rate=1)
    gradient_boosting_regressor.fit(x,y)
    subplot = ax[idx // 2][idx % 2]
    subplot.set_title("Griadient boosting model with n_estimator of 10 and Max_depth {}".format(max_dept))
    subplot.scatter(x,y)
    subplot.plot(x, gradient_boosting_regressor.predict(x), color='r')
plt.show()
```
