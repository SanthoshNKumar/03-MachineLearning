```python
df_0thDay1 = pd.read_csv("SP_123.csv")

print(df_0thDay1.shape)

X = df_0thDay1[['LOI950','SiO2','Al2O3','CaO','MgO','K2O','Na2O','Air','BLAINE']]

#X = df_0thDay1.drop(['SampleID','1CS'],axis=1)
y = df_0thDay1['1CS']

X = (X - np.mean(X))/np.std(X) # Standarize the data
```

```python
def get_plotofLearningVSIteration(X,y,lrate):
    
    # no. of training samples
    m = len(y) 
    ones = np.ones((m,1))
    X = np.hstack((ones, X))

    y = y[:,np.newaxis]

    theta = np.zeros(( X.shape[1],1))
    iteration = 1000

    loss = []
    iterations = []
    r2score = []

    for i in range(iteration):
        
        y_predicted = np.dot(X, theta)
        error = y_predicted - y
        sq_error =error**2
        cost = sq_error.sum() / (2*m)

        iterations.append(i)
        loss.append(cost)

        temp = y_predicted - y

        temp = np.dot(X.T, temp)
        theta = theta - (lrate/m) * temp
    return iterations,loss
```

```python
# Learning rate = 0.001
iterations,loss = get_plotofLearningVSIteration(X,y,0.0001)

plt.scatter(iterations,loss,label="Training")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
```
