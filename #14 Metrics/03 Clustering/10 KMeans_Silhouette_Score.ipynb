```python
# Find the Optimal Numbe of Clusters using silhouette_score

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
np.random.seed(100)

# Create a dataset
X,_= make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=1.5)
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.grid()
ax.set_xlabel("X",fontsize=30)
ax.set_ylabel("Y",fontsize=30)

ax.scatter(X[:,0],X[:,1],marker='o',color='b')
```

```python
nb_clusters = [2,3,4,5,6,7,8,9,10]

avg_silhouettee = []
```

```python
for i in nb_clusters:
    km = KMeans(n_clusters=i)
    Y= km.fit_predict(X)
    
    avg_silhouettee.append((silhouette_score(X,Y)))
```

```python
avg_silhouettee
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.plot(nb_clusters,avg_silhouettee,marker='o')
ax.grid(True)
ax.set_xlabel("Number of Clusters",fontsize=10)
ax.set_ylabel("Avg Silhouettee Score",fontsize=10)

# 3 is the best value
```
