```python
# Finding the Optimal number of Clusters using inertia

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
np.random.seed(100)

# Create a dataset
X,_= make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=1.5)
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.grid()
ax.set_xlabel("X",fontsize=30)
ax.set_ylabel("Y",fontsize=30)

ax.scatter(X[:,0],X[:,1],marker='o',color='b')
```

```python
inertias = []
```

```python
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit(X)
    inertias.append(km.inertia_)
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.grid()
ax.set_xlabel('Number of Cluster',fontsize=20)
ax.set_ylabel('Inertia',fontsize=20)
ax.plot(inertias,marker ='o',linewidth=2)
```
