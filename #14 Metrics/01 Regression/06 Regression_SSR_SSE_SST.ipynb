```python
# SSR : Sum of Square Regression
# SSE : Sum of Square Error
# SST : Sum of Square total

# SSR  = summation(y_mean - y_pred)
# SSE  = summation (y - y_pred)
# SST = summation(y - y_mean)  = SSR + SST
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({"X":[3,5,7,9,11,13],"y":[12,8,10,9,5,2]})

X = np.array(data['X']).reshape(-1,1)
y = np.array(data['y']).reshape(-1,1)

lr = LinearRegression()
lr.fit(X,y)

data['y_pred'] = lr.predict(X)

data
```

```python

y_mean = data['y'].mean()

print(y_mean)
```

```python
data['SSR:square(y_mean - y_pred)'] = np.square(y_mean - data['y_pred'])

data['SSE:square(y - y_pred)'] = np.square(data['y'] - data['y_pred'])

data['SST:square(y - y_mean)'] = np.square(data['y'] -y_mean)

data['SST: SSr + SSE'] = data['SSR:square(y_mean - y_pred)'] + data['SSE:square(y - y_pred)']
```
