```python
# L2 Regularization: Penalizes weights in portion to the sum of the squared of the weights.
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

plt.scatter(X,y)
```

```python
### Ridge Regression
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=1,solver="cholesky")
ridge_reg.fit(X,y)
predicted = ridge_reg.predict(X)

plt.scatter(X,y)
plt.scatter(X,predicted)
```

```python
### SGDRegressor with Pentalty terem of l2
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(penalty="l2")
sgd_reg.fit(X,y)
predicted = sgd_reg.predict(X)

plt.scatter(X,y)
plt.scatter(X,predicted)
```
