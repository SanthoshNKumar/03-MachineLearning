```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
```

```python
np.random.seed(1000)

X,y = make_classification(n_samples=500,n_features=4,n_informative=3,n_redundant=1,n_classes=3)
```

```python
 # Collect the scores for n_estimators in (1, 50)
a = []
max_estimates =50

for i in range(1,max_estimates):
    score = cross_val_score(GradientBoostingClassifier(n_estimators=i,learning_rate= 10.0/ float(i)),
                            X,y,cv=10,scoring='accuracy').mean()
    a.append(score)
    
```

```python
# Plot the results
plt.figure(figsize=(30, 25))
plt.xlabel('Number of estimators')
plt.ylabel('Average CV accuracy')
plt.grid(True)
plt.plot(a)
plt.show()
```
