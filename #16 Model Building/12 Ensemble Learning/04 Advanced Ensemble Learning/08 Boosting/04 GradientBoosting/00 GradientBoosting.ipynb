```python
'''
How Gradient Boosting Works:
Gradient boosting is another boosting model. Remember, boosting models keys is learning from the mistake.
It learns from the mistake by calculating residual error directly. Rather than update the weight of misclassified data points
```

```python
'''
How Gradient Boosting Differs from AdaBoost
    - Learning Mistake
       - AdaBoost Learns the Mistakes from update the weight of misclassified data points
       - Gradient Boosting learns the mistakes from calculating residual error directly
```

### Gradient Boosting Classifer

```python
import numpy as np
import pandas as pd

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv("pima-indians-diabetes.csv",names= names)
array = data.values
X = array[:,0:8]
Y = array[:,8]

seed = 7
num_trees = 100

from sklearn.model_selection import KFold
kfold = KFold(n_splits=10,random_state=seed)

from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=num_trees,random_state=seed)

from sklearn.model_selection import cross_val_score
results = cross_val_score(model,X,Y,cv = kfold)
print(results.mean())
```
