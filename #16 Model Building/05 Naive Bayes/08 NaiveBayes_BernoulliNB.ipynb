```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import cross_val_score

np.random.seed(1000)
X,y = make_classification(n_samples=300,n_features=2,n_informative=2,n_redundant=0)
```

```python
fig,ax = plt.subplots(1,1,figsize=(15,15))
ax.grid()
ax.set_xlabel("X",fontsize=30)
ax.set_ylabel("Y",fontsize=30)

# Plot the Lables
for i in range(len(X)):
    if y[i] ==0:
        ax.scatter(X[i,0],X[i,0],marker ='o',color ='r')
    else:
        ax.scatter(X[i,0],X[i,0],marker='^',color='b')
    
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)

# Create and train Bernoulli Naibe Bayes classifers
bnb = BernoulliNB(binarize=0.0)
bnb.fit(X_train,y_train)

print('Bernoulli Naive Bayes score: %.3f' % bnb.score(X_test, y_test))
```

```python
# Compute CV score
bnb_scores = cross_val_score(bnb, X, y, scoring='accuracy', cv=10)
print('Bernoulli Naive Bayes CV average score: %.3f' % bnb_scores.mean())
```

```python
# Predict some values
data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Yp = bnb.predict(data)
print(Yp)
```
