```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn.datasets import load_wine
dt = load_wine()
X = dt.data
y = dt.target
```

```python
independent_varibales = pd.DataFrame(dt.data,columns=dt.feature_names)

dependent_varibales = pd.Categorical.from_codes(dt.target,dt.target_names)
```

```python
winDataFrame = independent_varibales.join(pd.Series(dependent_varibales,name ="class"))
```

```python
winDataFrame.head()
```

```python
between_class_feature_means = pd.DataFrame(columns=dt.target_names)
for c,row in winDataFrame.groupby('class'):
    between_class_feature_means[c] = row.mean()
    
between_class_feature_means
```

```python
for c,row in winDataFrame.groupby('class'):
    print(row)
```

```python
within_class_scatter_matrix = np.zeros((13,13))
for c,row in winDataFrame.groupby('class'):
    rows = row.drop(['class'],axis=1)
    
    s = np.zeros((13,13))

    for index,row in row.iterrows():
        x,mc = row.values.reshape(13,1),between_class_feature_means[c].values.reshape(13,1)

        s += (x - mc).dot((x-mc).T)

    within_class_scatter_matrix += s

feature_means = winDataFrame.mean()
between_class_scatter_matrix = np.zeros((13,13))
for c in between_class_feature_means:
    n = len(winDataFrame.loc[winDataFrame['class'] ==c ].index)
    
    mc,m = between_class_feature_means[c].values.reshape(13,1),feature_means.values.reshape(13,1)
    
    within_class_scatter_matrix += n * (mc - m).dot((mc -m).T)
    
between_class_scatter_matrix
```

```python
eigen_values , eigen_vector = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix).dot(between_class_scatter_matrix))
```
