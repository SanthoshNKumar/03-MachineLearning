```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_wine
X, y = load_wine(return_X_y = True)

X.shape
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(24,14))
tree.plot_tree(clf, filled=True, fontsize=14)
```

```python
clf1 = tree.DecisionTreeClassifier(min_samples_leaf=5)
clf1.fit(X, y)

plt.figure(figsize=(24,14))
tree.plot_tree(clf1, filled=True, fontsize=14)
```

```python
# Name of the Attributes(Column) :  Since we did not name the columns, the index of the column is shown
#    - x[12]<=755.0
# Sample  Indicates that number of Observations(i.e rows) 
# Value shows the distribution of these samples according to the target variable
#     Values = [59,71,48]
#        - 59 = Class 1 :0
#        - 71 = Class 2 :1
#        - 48 = Class 3 :2
# If Gini = 0.0 means : samples are classified properly
```

```python
clf1 = tree.DecisionTreeClassifier(min_impurity_decrease=0.1)
clf1.fit(X, y)

plt.figure(figsize=(24,14))
tree.plot_tree(clf1, filled=True, fontsize=14)
```
