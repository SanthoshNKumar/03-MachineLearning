```python
'''
Isotonic Regression or Equal Stretch Regression
    - Isotonic Regression has flexibility in fitting monotonic functions better than Linear Regression
    - Greek root words: “iso” and “tonos“. “iso” meaning equal and “tonos” meaning to stretch
    - It works based on PAVA algorithm.
   
    
How PAVA Works
    - It inspects the points and if it finds a point that violates the constraints, it pools that value with its adjacent 
      members which ultimately go on to form a block.
      
      
1.It is a Piece wise Linear Model
2.It uses Monotonic constraints

- Draw thresholds with some equal intervals (Monotonics Constraints
      
1. [Initialization step] Set v and κ to any values satisfying conditions 1,3, and 4.
2. [Termination Step] If condition 2 (primal condition) is satisfied, then terminate.
3. [Pool Adjacent Violators] Choose any j such that νj > νj+1. And “pool” the blocks containing j and j + 1, that is, make
   them one block (and the nu or µ values for this pooled block will again be the weighted average of the values
   for this pooled block). [This step maintains conditions 1, 3, and 4.]
4. Go to Step 2 again.


**Note : A monotonic function is a function which is either entirely nonincreasing or nondecreasing.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.isotonic import IsotonicRegression
from matplotlib.collections import LineCollection
import random

n=100 #no. of points
x=np.arange(n)
y= 50*np.log(1+x) + np.random.randint(-50,50,size=(n,))

ir=IsotonicRegression()
lr=LinearRegression()
y_ir=ir.fit_transform(x,y)
lr.fit(x[:,np.newaxis],y)

lines=[[[i,y[i]],[i,y_ir[i]]] for i in range(n)]
lc=LineCollection(lines)

plt.figure(figsize=(16,4))
plt.plot(x,y,'r.',markersize=12)
plt.plot(x,y_ir,'b.-',markersize=12)
plt.plot(x,lr.predict(x[:, np.newaxis]), 'y-')
plt.gca().add_collection(lc)
plt.legend(('Data','Isotonic Fit','Linear Fit'))
plt.title("Isotonic Regression")
```
