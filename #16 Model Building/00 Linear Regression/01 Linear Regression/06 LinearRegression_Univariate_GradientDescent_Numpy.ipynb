```python
import numpy as np

def gradientDescent(x,y):    
    theta0=theta1 = 0
    iteration = 10
    n = len(x)
    learning_rate = 0.08
    print("***************************")
    print("Define Initial values:")
    print("Total Iteration",iteration)
    print("theta0 =",theta0)
    print("theta1 =",theta1)
    print("Learning rate =",learning_rate)
    print("n =",n)
    print("***************************")
    
    for i in range(iteration):
        print("Iteration = ",i)
        print("Step 1: Calculate the y_prediction")
        print("y_predicted = theta0 + theta1 * x")
        print("'''''''''''''")
        print("theta0 =", theta0)
        print("theta1 =", theta1)
        print("x =",x)
        print("theta0 + theta1 * x = ",theta0 + theta1 * x)
        y_predicted = theta0 + theta1 * x
        
        
        print("y_predicted = ",y_predicted)
    
        print("'''''''''''''")
    
        print("Step 2:Calculate Cost function")
        print("Cost= (1/n)Σ(y - y_predicted)²")
        
        print("'''''''''''''")
        print("y=",y)
        print("y_predicted = ",y_predicted)
        print("1. (y - y_predicted)²= ",[val**2 for val in (y-y_predicted)])
        print("2. Σ(y - y_predicted)² = ",sum([val**2 for val in (y-y_predicted)]))
        cost = (1/n)* sum([val**2 for val in (y-y_predicted)])
        print("3. (1/n)Σ(y - y_predicted)² = ",cost)
        print("Cost = ",cost)
        
        print("'''''''''''''")
        
        print("Step 3:Finding Gradient Descent")
        
        print("theta0 = theta0 - learning_rate * (-(2/n)*sum(y-y_predicted))")
        print("theta1 = theta1 - learning_rate * (-(2/n)*sum(x*(y-y_predicted)))")
        print("\n")
        
        print("Finding theta0")
        print("theta0 = theta0 - learning_rate * (-(2/n)*sum(y-y_predicted))")
        print("y =",y)
        print("y_predicted =",y_predicted)
        print("n =",n)
        print("\n")
        print("1. (y-y_predicted)) = ",y-y_predicted)
        print("2. sum(y-y_predicted)) = ",sum(y-y_predicted))
        print("3. (-(2/n)*sum(y-y_predicted))=",(-(2/n)*sum(y-y_predicted)))
        print("4. theta0 = theta0 - learning_rate * (-(2/n)*sum(y-y_predicted))")
        theta0 = theta0 - learning_rate * (-(2/n)*sum(y-y_predicted))
        print("5. theta0 = ",theta0)
        
        print("\n")
        
        print("Finding theta1")
        print("theta1 = theta1 - learning_rate * (-(2/n)*sum(x*(y-y_predicted)))")
        print("x =",x)
        print("y =",y)
        print("y_predicted =",y_predicted)
        print("n =",n)
        print("1. (y-y_predicted)) = ",y-y_predicted)
        print("2. x*(y-y_predicted)) = ",x*(y-y_predicted))
        print("3. sum(x*(y-y_predicted))) = ",sum(x*(y-y_predicted)))
        print("4. (-(2/n)*sum(x*(y-y_predicted))=",(-(2/n)*sum(x*(y-y_predicted))))
        
        theta1 = theta1 - learning_rate * (-(2/n)*sum(x*(y-y_predicted)))
        
        
        print("theta1 = ",theta1)
       
        print("##########################################################################################")    
        print("\n")

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradientDescent(x,y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ex1Data1.csv",header=None)

X = data.iloc[:,0] # Read First column
y = data.iloc[:,-1] # Read Second Column
m = len(y) # number of training example


X = X[:,np.newaxis]
y = y[:,np.newaxis]

theta = np.zeros([2,1])
iteration = 10
alpha =0.01
ones = np.ones((m,1))
X = np.hstack((ones,X))

def gradientDescent(X,y,theta,alpha,iterations):
    plt.figure(figsize=(20,10))
    plt.scatter(X[:,1], y)
    for i in range(iterations):
        temp = np.dot(X,theta) - y
        temp = np.dot(X.T,temp)
        theta = theta - (alpha / len(X)) * temp
        #print("GradientDescent for the Iteration :",i)
        #print(theta)
        #print('********')

        plt.plot(X[:,1], np.dot(X, theta),label= "Iteration_" + str(i))
        plt.legend()
        plt.show
    #print("*****************************************************")
    #print("Theta:\n",theta)
    #print("*****************************************************")
  
    return theta

theta1 = gradientDescent(X, y, theta, alpha, iteration)
```
