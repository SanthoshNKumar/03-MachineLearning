```python
'''
Sinusoidal Regression:
    - When the Data representation as like below (Sine waves) we go for Sinusoidal Regression
    - If your scatter plot shows a peak at x = 0, choose the cosine function.
    - If your scatter plot(Below Fig) is offset, with the peak closer to 1, choose the sine function.
```

```python
'''
formula:
Y = a*sin(b(X + c)) + d

a = amplitude
2Ï€/b is the period
c = phase shift
d = vertical shift.
```

```python
# Import required libraries:
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# Generating dataset:
# Y = A*sin(B(X + C)) + D
# A = Amplitude
# Period = 2*pi/B
# Period = Length of One Cycle
# C = Phase Shift (In Radian)
# D = Vertical Shift

X = np.linspace(0,1,100)             #(Start,End,Points)

# Hereâ€¦
# A = 1
# B= 2*pi
# B = 2*pi/Period
# Period = 1
# C = 0
# D = 0

Y = 1*np.sin(2*np.pi*X)
```

```python
plt.scatter(X,Y)# Plot the best fit curve :
plt.show()
```

```python
# Adding some Noise :
Noise = 0.4*np.random.normal(size=100)
Y_data = Y + Noise
plt.scatter(X,Y_data,c="r")
plt.show()
```

```python
# Function to calculate the value :
def calc_line(X,m,b):
  return b + X*m

# Calculate the value:
def calc_sine(x,a,b,c,d):
  return a * np.sin(b* ( x + np.radians(c))) + d
```

```python
# It returns optimized parametes for our function :
# popt stores optimal parameters
# pcov stores the covarience between each parameters.
popt,pcov = curve_fit(calc_line,X,Y_data)

# Plot the main data :
plt.scatter(X,Y_data)

# Plot the best fit line :
plt.title('Plotting Fit Line for Linear Regression')
plt.plot(X,calc_line(X,*popt),c="r")
plt.show()

# Check the accuracy of model :
Accuracy =r2_score(Y_data,calc_line(X,*popt))
print ("Accuracy of Linear Model : ",Accuracy)
```

```python
# Finding optimal parameters :
popt,pcov = curve_fit(calc_sine,X,Y_data)

plt.title('Plotting Fit Line for Sinusoidal Regression')

# Plot the main data :
plt.scatter(X,Y_data)# Plot the best fit curve :
plt.plot(X,calc_sine(X,*popt),c="r")
plt.show()

# Check the accuracy :
Accuracy =r2_score(Y_data,calc_sine(X,*popt))
print (Accuracy)
```
