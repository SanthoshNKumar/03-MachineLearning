```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

wine = load_wine()

X = wine.data
y = wine.target

neighbours = np.linspace(1,10,num=10)

accuracy = []

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(wine.data,wine.target)
    y_pred = knn.predict(wine.data)
    
    accuracy.append(accuracy_score(wine.target,y_pred))
    
    plt.plot(accuracy)
plt.xlabel("K Value")
plt.ylabel("Accuracy")
plt.title("KNN")
```

### Example 02

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()

X = iris.data[:,:2] #load the first two features of the iris data 
y = iris.target #load the target of the iris data

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=0)

all_scores= []

for i in range(3,9,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn,X_train,y_train,cv=10).mean()
    all_scores.append((i,score))
    
fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.grid()
ax.set_xlabel('Neighbors',fontsize=10)
ax.set_ylabel('Accuracy',fontsize=10)

ax.plot(np.array(all_scores)[:,0],np.array(all_scores)[:,1],marker='o',linewidth=2)
```

### Example 03

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Load datasets
wine = load_wine()

print(wine.feature_names)
print(wine.target_names)

# Split Dataset into training and Testing
X_train,X_test,y_train,y_test = train_test_split(wine.data,wine.target,test_size=0.3)

scores = {}
score_list = []

k_range = range(1,26)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test,y_pred)
    score_list.append(metrics.accuracy_score(y_test,y_pred))
    print("Accuracy is ", metrics.accuracy_score(y_test,y_pred)*100,"% for K-Value:",k)
    
plt.plot(k_range,score_list)
plt.xlabel("Value of K for KNN")
plt.ylabel("Testing Accuracy")
```
