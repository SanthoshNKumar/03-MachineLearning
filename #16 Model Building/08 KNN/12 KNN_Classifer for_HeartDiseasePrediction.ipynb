```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')

df.head()
```

```python
df['target'].value_counts().plot(kind='bar',color='green')
#Total number of hart disease predicted Postive or Negative"
```

```python
df.corr()
```

```python
# It is always to good to check the dataset whete the target classes are of approximately equal size(Check for imbalanced data)

sns.countplot(x='target',data=df)
```

```python
from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler()
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
df_scaled  = standardscaler.fit_transform(df[columns_to_scale])
```

```python
df.drop(columns_to_scale,axis=1,inplace=True)

df = pd.concat([df,pd.DataFrame(df_scaled,columns=columns_to_scale)],axis=1)

X = df[df.columns.difference(['target'])] # Slect all columns except 'Target'

y = df.target # target is your final output classifer
```

```python
from sklearn.model_selection import train_test_split

X_tran,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)

# Lets try KNN first

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

accuracy = []

for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    
    model = knn.fit(X_tran,y_train)
    
    y_pred = model.predict(X_test)
    
    accuracy.append(accuracy_score(y_pred,y_test).mean())
    
plt.plot([k for k in range(1,21)],accuracy,color='red')
for i in range(1,21):
    plt.text(i,accuracy[i-1],(i,accuracy[i-1]))
plt.xlabel("Number of Neighbors(k)")
plt.ylabel('Scores')
```

```python
# We can consider K value as 6

from sklearn.model_selection import cross_val_score
knn_classifer = KNeighborsClassifier(n_neighbors=6)
model = knn_classifer.fit(X_tran,y_train)
score = cross_val_score(knn_classifer,X_tran,y_train,cv=10)
accuracy.clear()
accuracy.append(score.mean())
```
