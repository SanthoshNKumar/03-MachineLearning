```python
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Assigning features and label variables
# First feature
weather=['Sunny','Sunny','Overcast','Rainy',
         'Rainy','Rainy','Overcast','Sunny',
         'Sunny','Rainy','Sunny','Overcast',
         'Overcast','Rainy']

#Second Feature
temp=['Hot','Hot','Hot','Mild','Cool',
      'Cool','Cool','Mild','Cool',
      'Mild','Mild','Mild','Hot','Mild']

# Label or target varible
play=['No','No','Yes','Yes',
      'Yes','No','Yes','No',
      'Yes','Yes','Yes','Yes',
      'Yes','No']

pd.DataFrame({"Weather": weather,"Temprature":temp,"Play":play})
```

```python
# Step 1: Encoding Data Columns
#        - Weater
#        - Temperature
#        - Play

#creating labelEncoder
le = preprocessing.LabelEncoder()


# Converting string labels into numbers.
weather_encoded=le.fit_transform(weather)

# converting string labels into numbers
temp_encoded=le.fit_transform(temp)

# Convert string labels into numbers
label_encoded=le.fit_transform(play)


print(weather_encoded)

#combinig weather and temp into single listof tuples
features=list(zip(weather_encoded,temp_encoded))

# Step 2: Apply KNN
model = KNeighborsClassifier(n_neighbors=3)
model.fit(features,label_encoded)
y_pred = model.predict([[2,2]])
```
