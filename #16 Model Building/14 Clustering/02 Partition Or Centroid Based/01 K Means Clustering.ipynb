```python
'''
K Means Clustering:
    -The goal of this algorithm is to find groups in the data, with the number of groups represented by the variable K.
    -K-means clustering is one of the simplest and popular unsupervised machine learning algorithms.
    -A cluster refers to a collection of data points aggregated together because of certain similarities.
    -The ‘means’ in the K-means refers to averaging of the data; that is, finding the centroid
    
The algorithm works iteratively to assign each data point to one of K groups based on the features that are provided. 
Data points are clustered based on feature similarity. The results of the K-means clustering algorithm are:
    -The centroids of the K clusters, which can be used to label new data
    -Labels for the training data (each data point is assigned to a single cluster)

List of Application using K means Clustering:
    -Group customer based on criteria’s such as willingness, purchasing power etc.
    -Clustering is used to study earth-quake. Based on the areas hit by an earthquake in a region, clustering can help analyses 
     the next probable location where earthquake can occur.
    -Document Classification based on tags, title and other contents of the document.
    -Insurance Fraud Detection
```

```python
'''
Algorithm:
    The Κ-means clustering algorithm uses iterative refinement to produce a final result. 
    The algorithm inputs are the number of clusters Κ and the data set. The data set is a collection of features for each data 
    point. 
    The algorithms starts with initial estimates for the Κ centroids, which can either be randomly generated or randomly 
    selected from the data set.
The algorithm then iterates between two steps:
    -Data assignment step:
    -Centroid update step:

How the Algorithms Works?
    1.First, we initialize k points, called means, randomly.
    2.We categorize each item to its closest mean and we update the mean’s coordinates, which are the averages of the items 
      categorized in that mean so far.
    3.We repeat the process for a given number of iterations and at the end, we have our clusters.
```

### Example 01

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

X =  np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])
X = pd.DataFrame(X,columns=['Column1','Column2'])
plt.scatter(X['Column1'],X['Column2'])

kmeans = KMeans(n_clusters=2,random_state=0).fit(X)
kmeans.labels_
kmeans.predict([[0,0],[12,3]])
kmeans.cluster_centers_
```

### Example 02

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])

plt.plot()
plt.xlim([0,10])
plt.ylim([0,10])

plt.xlabel("x1")
plt.xlabel("x2")
plt.scatter(x1,x2)

# create new plot and data
plt.plot()
X = np.array(list(zip(x1,x2))).reshape(len(x1),2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']

#K means Clustring
k =3
kmeans_model = KMeans(n_clusters=k).fit(X)
plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    
plt.show()
```

### Example 03

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

np.random.seed(100)

# Create a dataset
X,_= make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=1.5)

fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.grid()
ax.set_title("Actual Data")
ax.set_xlabel("X",fontsize=30)
ax.set_ylabel("Y",fontsize=30)

ax.scatter(X[:,0],X[:,1],marker='o',color='b')

# Create and Train K-Means
km = KMeans(n_clusters=3)
km.fit(X)

# Show the centroids
print("Custer Centroids: \n",km.cluster_centers_)

fig,ax = plt.subplots(1,1,figsize=(7,5))
ax.set_title("Grouped Data Points")
ax.grid()
ax.set_xlabel('X',fontsize=30)
ax.set_ylabel('Y',fontsize=30)

for i in range(1000):  # nb_samples = 1000
    c = km.predict(X[i].reshape(1,-1))
    if c==0:
        ax.scatter(X[i,0],X[i,1],marker='o',color='r')
    elif c==1:
        ax.scatter(X[i,0],X[i,1],marker='^',color='b')
    else:
        ax.scatter(X[i,0],X[i,1],marker='d',color='g')
plt.show()
```

### Example 04

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans

# for Reproducibility
np.random.seed(1000)

# Create Dataset
X,y = make_circles(n_samples=1000,noise=0.05)

fig,ax = plt.subplots(1,1,figsize=(7,5))

ax.grid()
ax.set_xlabel("X",fontsize=30)
ax.set_ylabel("y",fontsize=30)

for i in range(len(X)):
    if y[i] ==0:
        ax.scatter(X[i,0],X[i,1],color='r')
    else:
        ax.scatter(X[i,0],X[i,1],color='b')
        
# Create and train K-Means
km=KMeans(n_clusters=2)
km.fit(X)

# Show clustered dataset
fig,ax = plt.subplots(1,1,figsize=(7,5))

ax.set_xlabel('X',fontsize=30)
ax.set_ylabel('Y',fontsize=30)

for i in range(1000):  # nb_samples = 1000
    c = km.predict(X[i].reshape(1,-1))
    if c==0:
        ax.scatter(X[i,0],X[i,1],marker='o',color='r')
    else :
        ax.scatter(X[i,0],X[i,1],marker='^',color='b')
plt.show()
```

### Example 05: K Means_Elbow Method

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn.datasets import make_blobs
np.seterr(divide = 'ignore') 

#Prepare DataSet
X,y_true = make_blobs(n_samples=300,centers =4,cluster_std=0.60,random_state=0)

plt.scatter(X[:,0],X[:,1],s =50)
plt.show()

kvalues = []
errors = []

for i in range(1,5):
    kmeans = cluster.KMeans(n_clusters=i)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    error = ((y_true - y_kmeans)**2).sum()
    errors.append(error)
    kvalues.append(i)
    
print(kvalues)
print(errors)

plt.plot(kvalues,errors)
plt.scatter(kvalues,errors)
plt.ylabel("Sum of Mean Square Error")
plt.xlabel("K value")
plt.title("Elbow Method and Finding the Right Number of Clusters")
```
