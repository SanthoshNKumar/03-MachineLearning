```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("concrete_data.csv")
```

```python
X = df.drop(['concrete_strength'],axis=1).values

y = df['concrete_strength'].values.reshape(-1,1)
```

```python
X = np.concatenate((np.ones((len(X),1)),X),axis=1)
X.shape
```

```python
Xnorm = X.copy()

minx = np.min(X[:,1:])
maxx = np.max(X[:,1:])
Xnorm[:,1:] = (X[:,1:]-minx)/(maxx-minx)
```

```python
theta0 = np.zeros((X.shape[1],1))+0.4

# Momentum Gradient Descent

m = len(X)
epochs = 1000
gamma = 0.1
eta = 0.1

v_b,v_w = 0,0

for i in range(epochs):
    
    print("*********************{0}**************************".format(i))
    
    cost = np.sum((np.dot(Xnorm,theta0)-y)**2,axis=0)[0]
    print("Cost:",cost)
    
    # Derivatie of Loss function
    dtheta0 = 1/m*np.sum((Xnorm.dot(theta0)-y)*Xnorm,axis=0).reshape(-1,1)
    
    # b = theta0[:1]
    # db = dtheta0[:1]
    # v_b = gamma * v_b + eta * db

    # w = theta0[1:]
    # dw = dtheta0[1:]
    #v_w = gamma * v_w + eta * dw

    # w = w - v_w
    # b = b - v_b
    
    # Update Weights
    v_b = gamma * v_b + eta * theta0[:1]
    v_w = gamma * v_w + eta * theta0[1:]

    theta0[:1] = theta0[:1] - v_b
    theta0[1:] = theta0[1:] - v_w
    
```
