```python
'''
Why ONNX ?
    1. PyTorch, Tensorflow, Caffe2, MXNet, etc, are just some of the most popular frameworks today for the development 
       of Deep Learning models.
    2. Some frameworks are more research-oriented (such as PyTorch) and some others are mostly used for device deployment 
       (such as Tensorflow)
    3. Obviously, each of these frameworks offers different advantages over the others.
    4. How could we interoperate different frameworks?
    5. How could we optimize a model with framework “x” and deploy it in an architecture optimized for framework “y”?
    6. This type of interoperability is achieved thanks to the ONNX standard and the ONNX Runtime.
    7. ONNX has come to break the dependency between frameworks and hardware architectures.

'''
```
